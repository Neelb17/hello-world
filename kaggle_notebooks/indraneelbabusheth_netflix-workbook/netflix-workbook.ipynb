```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

netflix_df = pd.read_csv('/kaggle/input/netflix-shows/netflix_titles.csv')
print(netflix_df)




#counting titles per genre
G_list = netflix_df["listed_in"].str.split(", ")
flat_genre = sum(G_list, [])
C_Genre = pd.Series(flat_genre).value_counts()
#print(C_Genre)

#title released in my birth year
netflix_df["Date_Variable"] = pd.to_datetime(netflix_df["date_added"] , errors="coerce")
netflix_df["Title_Year"] = netflix_df["Date_Variable"].dt.year.astype("Int64")
Y_1996 = netflix_df["Title_Year"] == 2015
F_output = netflix_df[Y_1996][["show_id","title","Title_Year"]]
print(F_output)

#Title with larget cast list

netflix_df["Cast_List"] = netflix_df["cast"].str.split(", ")
netflix_df["Count_CastList"] = netflix_df["Cast_List"].dropna().apply(len)
highest_cast = netflix_df["Count_CastList"].idxmax()
all_att = netflix_df.loc[highest_cast,["title", "cast", "Count_CastList"]]
print(all_att)


#Titles by Duration Type"
netflix_df["d_type"] = netflix_df["duration"].str.extract("(min|Season|Seasons)", expand=False)

count_d_type = netflix_df["d_type"].value_counts().sort_values(ascending=False)
print(count_d_type)

#Title with no cast information
no_cast = netflix_df["cast"].isna()
title_wncast = netflix_df[no_cast][["show_id","title","cast"]]
print(title_wncast)

#Movies by Decade
netflix_df["Decade"] = (netflix_df["Title_Year"] // 10) * 10

decade_counts = netflix_df["Decade"].value_counts().sort_index

print(decade_counts)

#Finding keyword in desc
crime_desc = netflix_df["description"].str.contains("crime", case = False, na=False)
all_atwc = netflix_df[crime_desc][["title","type","description"]]
print(all_atwc)

#Shortest & Longest Tile
netflix_df["title_len"] = netflix_df["title"].str.len()

shortest_title = netflix_df["title_len"].idxmin()
longest_title = netflix_df["title_len"].idxmax()

short_df = netflix_df.loc[[shortest_title], ["title", "type", "title_len"]]
long_df = netflix_df.loc[[longest_title], ["title", "type", "title_len"]]

print(short_df)
print(long_df)

#% of missing values

missing_values = netflix_df.isna().mean() * 100
print(missing_values)

#most common release year for movies
t_movie = netflix_df[netflix_df["type"] == "Movie"]
count_ryear = t_movie["Title_Year"].value_counts()


most_common_year = count_ryear.idxmax()
count_value = count_ryear.max()

print(most_common_year)
print(count_value)

#No. of Season
TV_Show_F = netflix_df[netflix_df["type"] == "TV Show"]
netflix_df["Ex_Season"] = TV_Show_F["duration"].str.extract(r'(\d+)')
ct_of_seasons = netflix_df["Ex_Season"].value_counts()

print(ct_of_seasons)

#Titles with more than 3 genre
netflix_df["genre_list"] = netflix_df["listed_in"].str.split(", ")
netflix_df["genre_count"] = netflix_df["genre_list"].apply(len)
mt3_genre = netflix_df["genre_count"].idxmax()

Title_mt3g = netflix_df.loc[mt3_genre,["title","listed_in","genre_count"]]
print(Title_mt3g)

#Top 10 Directors with most titles
netflix_df["Director_List"] = netflix_df["director"].str.split(", ")
final_list_d = sum(netflix_df["Director_List"].dropna(), [])
count_of_Director = pd.Series(final_list_d).value_counts().head(10)
print(count_of_Director)
```
